# LISTING CLEANING

```python
with open('barcelona_listings_splited.csv') as f:
    cf = csv.reader(f)
    writer = csv.writer(open("barcelona_listings_splited_cleaned_withNone.csv", 'a'))
    
    max_len = 0
    sr = ''
    id_max=''
    for row in cf:
        new_row = list()
        
        def stringCleanerAppender(rowNumber, endChar):
            clean_string = (row[rowNumber][:endChar]).replace('\n', '').replace('|', '')
            if(clean_string==''):
                new_row.append(None)
            else:
                new_row.append(clean_string)
                
            
        def digitAppender(rowNumber):
            if row[rowNumber].isdigit():
                new_row.append(row[rowNumber])
            else:
                new_row.append(None)
            
        #check id
        if row[0].isdigit():
            new_row.append(row[0])
        else: 
            continue
            
    
        #correct url
        #row[1]
        #new row[1]
        url = 'https://www.airbnb.com/rooms/' + str(row[0])
        new_row.append(url)
       
    
        #name
        stringCleanerAppender(2, 90)
        #summary
        stringCleanerAppender(3, 1000)
        #space
        stringCleanerAppender(4, 1000)
        #description
        stringCleanerAppender(5, 1000)
        #neighborhood overview
        stringCleanerAppender(6, 1000)
        #notes
        stringCleanerAppender(7, 1000)
        #transit
        stringCleanerAppender(8, 1000)
        #access
        stringCleanerAppender(9, 1000)
        #interaction
        stringCleanerAppender(10, 1000)
        #house_rules
        stringCleanerAppender(11, 1000)
        
        #correct picture url
        #row[12] 
        if 'https://' in row[12]:
            stringCleanerAppender(12, 120)
        else: 
            new_row.append(None)
            
        #host id
        #if correct host
        if row[13].isdigit():
            new_row.append(row[13])

            #host url
            host_url = 'https://www.airbnb.com/users/show/' + str(row[13])
            new_row.append(host_url)
            #host name
            stringCleanerAppender(15, 35)

            #host since
            if row[16]=='':
                new_row.append('')
            elif row[16][0]=='2' and (
                row[16][1]=='0') and (
                row[16][2].isdigit()) and (
                row[16][3].isdigit()) and (
                row[16][4]=='-') and (
                row[16][5].isdigit()) and (
                row[16][6].isdigit()) and (
                row[16][7]=='-') and (
                row[16][8].isdigit()) and (
                row[16][9].isdigit()):
                new_row.append(row[16])
            else:
                new_row.append(None)
                
            #host about
            stringCleanerAppender(17, 5730)
            
            #host response rate
            i=0
            rate=''
            while row[19][i].isdigit():
                rate+=str(row[19][i])
                i+=1
            if(rate==""):
                new_row.append(None)
            else:
                new_row.append(rate)
            
            #host thumbnail url
            if 'https://' in row[20]:
                stringCleanerAppender(20, 200)
            else: 
                new_row.append(None)
                
            #host picture url
            if 'https://' in row[21]:
                stringCleanerAppender(21, 200)
            else: 
                new_row.append(None)
        
        #if host without id    
        else:
            continue
                
        #neighbourhood
        stringCleanerAppender(24, 50)
        
        #city
        new_row.append('Barcelona')
        #country code
        new_row.append('ES')
        #country
        new_row.append('Spain')
            
        #latitude and longitude
        #je les fais d'un coup parce que si l'un l'est et pas l'autre à 
        #99% c'est un décalage donc mauvaises données
        if isfloat(row[28]) and isfloat(row[29]):
            new_row.append(row[28])
            new_row.append(row[29])
        else:
            new_row.append(None)
            new_row.append(None)
        
        #accomodates
        digitAppender(32)
            
        #bathrooms
        digitAppender(33)
            
        #bedrooms
        digitAppender(34)
        
        #beds
        digitAppender(35)
        
        #square feet 
        digitAppender(38)
        
        def priceAppender(rowNumber):
            if row[rowNumber]=='':
                new_row.append(None)
            elif row[rowNumber][0]=='$':
                i=1
                pr=''
                while row[rowNumber][i].isdigit():
                    pr+=row[rowNumber][i]
                    i+=1
                new_row.append(pr)
            else:
                new_row.append(None)
            
        
        #price
        priceAppender(39)
        
        #weekly price
        priceAppender(40)
        
        #monthly price
        priceAppender(41)
            
        #security deposit
        priceAppender(42)
        
        #cleaning fee
        priceAppender(43)
            
        #guests included
        digitAppender(44)
        
        #extra people
        priceAppender(45)
            
        #minimum nights
        digitAppender(46)
            
        #maximum nights
        digitAppender(47)
            
        #review score rating
        digitAppender(48)
        
        #review scores accuracy
        digitAppender(49)
        
        #review scores cleanliness
        digitAppender(50)
        
        #review scores checking
        digitAppender(51)
        
        #review scores communication
        digitAppender(52)
        
        #review scores location
        digitAppender(53)
        
        #review scores value
        digitAppender(54)
            
            
        #is business travel ready
        if not (row[55]=='t' or row[55]=='f'):
            new_row.append(None)
        
        new_row.append(row[55])
        
        
        #require guest profile picture + phone verification
        if (row[57]=="t" or row[57]=="f") and (row[58]=="t" or row[58]=="f"):
            new_row.append(row[57])
            new_row.append(row[58])
            writer.writerows([new_row])
        

        
        
    
```
